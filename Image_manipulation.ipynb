```python
def plti(camera, h = 8, **kwargs):
    y = camera.shape[0]
    x = camera.shape[1]
    w = (y/x) * h
    plt.figure(figsize = (w,h))
    plt.imshow(camera, interpolation="none", **kwargs)
    plt.axis('off')

plti(camera)
```

## Converting to grayscale

```python
def to_grayscale(camera, weights = np.c_[0.2989, 0.5870, 0.1140]):
    tile = np.tile(weights, reps=(camera.shape[0], camera.shape[1],1))
    return np.sum(tile * camera, axis=2)

    camera = to_grayscale(camera)

plti(camera, cmap='gray')
```
