### Image editing using numpy

Since Images are also arrays, we can use numpy to edit them

```python
from skimage import data
camera = data.camera()

from matplotlib import pyplot as plt
plt.imshow(camera, cmap='gray'); # Use this function to display the image after editing
```

```python
camera = data.camera()
fig = plt.figure()
a = fig.add_axes([0.1, 0.1, 0.8, 0.8])
b = fig.add_axes([0.65, 0.55, 0.1, 0.3]) 
a.imshow(camera,cmap='gray');
#a.grid(True)
x=camera[60:160, 170:270]
x=np.fliplr(x)
b.imshow(x, cmap='gray'); 
b.axis('off')
```
