### Image editing using numpy

Since Images are also arrays, we can use numpy to edit them

```python
%matplotlib inline
```

```python
from skimage import data
camera = data.camera()

from matplotlib import pyplot as plt
plt.imshow(camera, cmap='gray'); # Use this function to display the image after editing
```

Use numpy functions to produce the following image

https://i.ibb.co/1zf8dvC/index.png

![Result](https://i.ibb.co/1zf8dvC/index.png)

```python
import numpy as np
camera.shape
```

```python
type(camera)
```

```python
x = camera[70:170,180:280]
plt.imshow(x, cmap='gray')
plt.show()
```

```python
x_flipped = np.fliplr(x)
plt.imshow(x_flipped, cmap='gray')
plt.show()
```

```python
camera[100:200, 380:480] = x_flipped

plt.imshow(camera, cmap='gray')
plt.show()
```
