```python
#library imports
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
```

```python
# standardization of the data
def stand(df):
    for i in df.columns:
        if i=="strength": 
            break
        scale = StandardScaler().fit(df[[i]])
        df[i] = scale.transform(df[[i]])
        return df
```

```python
#train data import and standardization
df = stand(pd.read_csv('train_data.csv'))

```

```python
#data splits
dataset = df.values
dataset
X = dataset[:,0:8]

Y = dataset[:,8]
```

```python
#model
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(512, activation='relu'))
model.add(tf.keras.layers.Dense(64, activation='relu'))
model.add(tf.keras.layers.Dense(32, activation='relu'))
model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dense(512, activation='relu'))
model.add(tf.keras.layers.Dense(64, activation='relu'))
model.add(tf.keras.layers.Dense(32, activation='relu'))
model.add(tf.keras.layers.Dense(128, activation='relu'))
model.add(tf.keras.layers.Dense(1))
```

```python
from tensorflow.keras.optimizers import *

model.compile(optimizer=Adam(lr=0.001),loss="mean_squared_error")
class myCallback(tf.keras.callbacks.Callback):
    def on_epoch_end(self,epoch,logs={}):
        if(logs.get("loss")<1.95):
            self.model.stop_training=True
callbacks=myCallback()
```

```python
model.fit(X,Y,epochs=3000,shuffle=True,callbacks=[callbacks])
```

```python
#predictions for train
prediction = model.predict(X[:])
y_predicted=[]
for i in range(len(prediction)):
    y_predicted.append(prediction[i][0])
y_actual=Y
```

```python
#Root Mean Squared Error
from sklearn.metrics import mean_squared_error
from math import sqrt
rms = sqrt(mean_squared_error(y_actual, y_predicted))
print(rms)
```

```python
#test data import and standardization
df2=stand(pd.read_csv('test_data.csv'))
```

```python
#predictions for test
X=df2.values
prediction = model.predict(X[:])
y_predicted=[]
for i in range(len(prediction)):
    y_predicted.append(prediction[i][0])
```

```python
#saving predictions of test to csv
a=[i for i in range(0,len(df2)+1)]
df3=pd.DataFrame(list(zip(a, y_predicted)),columns =['Id', 'Predicted'])
df3.to_csv("submission.csv",index=False)
```
